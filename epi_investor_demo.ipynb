# üëÅÔ∏è VIEW: Interactive Timeline

See **every AI decision** in an interactive interface


```python
# @title üëÅÔ∏è View Authentic EPI Viewer { display-mode: "form" }
import zipfile, json, html, os
from pathlib import Path
from IPython.display import display, HTML

epi_files = list(Path('.').glob('*.epi')) + list(Path('.').glob('epi-recordings/*.epi'))
epi_file = max(epi_files, key=os.path.getmtime) if epi_files else None

if epi_file:
    print("=" * 70)
    display(HTML('<h2 style="color: #3b82f6;">üëÅÔ∏è Extracting authentic EPI viewer...</h2>'))
    
    signature_id = "UNSIGNED"
    viewer_html = None
    
    try:
        with zipfile.ZipFile(epi_file, 'r') as z:
            if 'manifest.json' in z.namelist():
                m = json.loads(z.read('manifest.json').decode('utf-8'))
                s = m.get('signature', '')
                if ':' in s:
                    signature_id = f"{s.split(':')[0].upper()}...{s.split(':')[-1][:16]}"
            
            for fname in z.namelist():
                if fname.endswith('.html'):
                    viewer_html = z.read(fname).decode('utf-8', errors='ignore')
                    break
    except Exception as e:
        print(f"Error: {e}")
    
    if viewer_html:
        escaped = html.escape(viewer_html)
        iframe = f'<div style="border: 4px solid #10b981; border-radius: 16px; overflow: hidden; margin: 25px 0;"><div style="background: linear-gradient(135deg, #10b981, #059669); color: white; padding: 18px 24px; display: flex; justify-content: space-between; align-items: center;"><span style="font-size: 22px; font-weight: bold;">üõ°Ô∏è AUTHENTIC EPI VIEWER</span><span style="font-family: Courier New, monospace; font-size: 14px; background: rgba(255,255,255,0.25); padding: 8px 14px; border-radius: 8px;">{signature_id}</span></div><iframe srcdoc="{escaped}" width="100%" height="700" style="border: none;" sandbox="allow-scripts allow-same-origin"></iframe></div>'
        
        print("=" * 70)
        display(HTML('<h1 style="color: #10b981; font-size: 36px; margin: 20px 0;">‚úÖ AUTHENTIC VIEWER LOADED</h1>'))
        print("This is the REAL EPI viewer from the .epi file")
        print("=" * 70)
        display(HTML(iframe))
    else:
        display(HTML('<h2 style="color: #ef4444;">‚ùå No viewer.html found</h2>'))
else:
    print("Run demo cell first")
```

**Let's try to create fraudulent evidence and see if EPI catches it.**
